```python
import spacy
import torch
import pandas as pd
from torch.nn.utils.rnn import pad_sequence

# Load spaCy's medium-sized model with GloVe vectors (300D)
nlp = spacy.load('en_core_web_md')

# Example: A list of text sentences
sentences = pd.Series([
    "I love natural language processing.",
    "PyTorch is a great framework for deep learning.",
    "GloVe embeddings help in understanding word meanings."
])

# Function to convert each sentence into a list of word vectors
def sentence_to_word_vectors(sentence, model):
    doc = model(sentence)  # Tokenize sentence using spaCy
    word_vectors = [token.vector for token in doc]  # Get vectors for each word
    return torch.tensor(word_vectors)  # Convert list of vectors to a tensor

# Convert each sentence to a sequence of word vectors
sentence_word_vectors = [sentence_to_word_vectors(sentence, nlp) for sentence in sentences]

# Pad the sequences to make them of the same length (sequence_len)
# The padding is necessary because sentences have different lengths
padded_word_vectors = pad_sequence(sentence_word_vectors, batch_first=True)

# Check tensor shape (batch_size, sequence_len, embed_dim)
print(f"Tensor shape: {padded_word_vectors.shape}")
```
