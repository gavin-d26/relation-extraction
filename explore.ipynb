```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/hw1_train.csv")
```

```python
df["CORE RELATIONS"] = df["CORE RELATIONS"].str.split(" ")
```

```python
# df["CORE RELATIONS"]=df["CORE RELATIONS"].apply(lambda x: sorted(x))
```

```python
edf = df.explode('CORE RELATIONS')
```

```python
edf["values"]=str(1)
```

```python
pdf=edf.pivot(columns="CORE RELATIONS", index='ID', values='values').fillna(str(0))
```

```python
pdf
```

```python
pdf['UTTERANCES'] = df['UTTERANCES']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, Y = pdf.iloc[:, 0:1], pdf.iloc[:, 1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.10, random_state=0)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
val_df = pd.concat([X_test, y_test], axis=1)
```

```python
train_df.to_csv("data/train_df.csv", index=True)
val_df.to_csv("data/val_df.csv", index=True)
```
